## Importing required Modules

```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression 
```

## Reading table using Panda and its description

```python
data = pd.read_csv('cost_revenue_clean.csv')
data
```

```python
data.describe()
```

## Feeding data to the axes and plotting it using matplotlib

```python
X = DataFrame(data, columns=['production_budget_usd'])
Y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,Y, alpha=0.1)
plt.title('Movie Budget VS Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Global Gross Revenue')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

## Calculating Slope and Intercept using Linear Regression

```python
lr = LinearRegression()
lr.fit(X, Y)
```

```python
lr.coef_ #slope cofficient
```

```python
lr.intercept_ #slope intercept
```

## Plotting the line using Matplotlib

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,Y, alpha=0.1)
plt.plot(X, lr.predict(X), color='red', linewidth=3)
plt.title('Movie Budget VS Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Global Gross Revenue')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

## Calculating " GOODNESS OF FIT "

```python
lr.score(X,Y)
```
